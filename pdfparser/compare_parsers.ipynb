```python
import pdfplumber
import pandas as pd

pd.set_option('display.max_colwidth', None)


fname = "../data/Workstation monitor_24003PPT.pdf"
pdf = pdfplumber.open(fname)
# pdf_length = len(pdf.pages)


for page_index, page in enumerate(pdf.pages):
    page_text = page.extract_text()
    tables = page.extract_tables(
        table_settings = {
            "vertical_strategy": "lines", 
            "horizontal_strategy": "lines",
        }
    )
    print(f'/----------------------------------------------------- Page {page_index+1} Start -----------------------------------------------------\\')
    print(f'-----------------------------------------------------  Raw Text  -----------------------------------------------------')
    print(page_text)
    print(f'-----------------------------------------------------English Text-----------------------------------------------------')
    print()
    print(f'----------------------------------------------------- Korean Text-----------------------------------------------------')
    
    print()
    print(f'--------------------------------------------------- Summarization ---------------------------------------------------')
    print()
    for table_index, table in enumerate(tables):
        print(f'-----------------------------------------------------  Raw Table  -----------------------------------------------------')
        print(f'Table {page_index+1}-{table_index+1}')
        # Convert list of lists to dataframe
        df = pd.DataFrame(table)
        print(df)
        print()
        # print(f'-----------------------------------------------------English Table-----------------------------------------------------')
        # print()
        # print(f'----------------------------------------------------- Korean Table-----------------------------------------------------')
        # print()
    print(f'\\------------------------------------------------------ Page {page_index+1} End ------------------------------------------------------/')
    print()
```

```python
import fitz

doc = fitz.open("../data/Workstation monitor_24003PPT.pdf") # open a document
out = open("../data/output.txt", "wb") # create a text output
for page in doc: # iterate the document pages
	text = page.get_text().encode("utf8") # get plain text (is in UTF-8)
	print(text)
	out.write(text) # write text of page
	out.write(bytes((12,))) # write page delimiter (form feed 0x0C)
out.close()
```
